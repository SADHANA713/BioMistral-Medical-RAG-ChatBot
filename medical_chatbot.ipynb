#Importing Libraries

```python
from langchain.document_loaders import PyPDFDirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.embeddings import SentenceTransformerEmbeddings
from langchain.vectorstores import Chroma
from langchain_community.llms import LlamaCpp
from langchain.chains import RetrievalQA,LLMChain
```

#import the Document

```python
loader =PyPDFDirectoryLoader("/content/drive/MyDrive/medical chatbot/")
docs=loader.load()
```

#chunking

```python
test_splitter=RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50 )
chunks=test_splitter.split_documents(docs)
```

```python
embeddings = SentenceTransformerEmbeddings(model_name="NeuML/pubmedbert-base-embeddings")
```

#Vectore Store CCreation

```python
vectorstore = Chroma.from_documents(chunks, embeddings)
```

```python
query = "who is at risk of heart disease?"

search_results = vectorstore.similarity_search(query)
```

```python
search_results
```

```python
retriever=vectorstore.as_retriever(search_kwags={'k':5})
```

```python
retriever.get_relevant_documents(query)
```

#LLM MODEL LOADING

```python
llm = LlamaCpp(
    model_path="/content/drive/MyDrive/medical chatbot/BioMistral-7B.Q4_K_S.gguf",
    temperature=0.2,
    max_tokens=2048,
    top_p=1
)
```

#use llm and retriver and query ,to generate final response

```python
template ="""
<|context|>
Your are Medical Assistant that follows the instructions and generate the accurate response based on the query and the context provided
Please be truthful and give direc answers.
</s>
<|user|>
{query}
<|assistant|>
"""
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
from langchain.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
rag_chain =(
    {"conext":retriever,"query":RunnablePassthrough()}
    |prompt
    |llm
    |StrOutputParser()
)
```

```python
response = rag_chain.invoke(query)
```
